```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

userDfFull = pd.read_csv("UserInfo.csv")

userDfFull.head()
```

```python
redemptions = pd.read_csv("Redemptions.csv")
redemptions['date'] = pd.to_datetime(redemptions['timestamp'], unit='s')

redemptions.head()
```

```python
minTVL = 10000
userDf = userDfFull[userDfFull.Value > minTVL]
```

```python
otherApps = list(userDf['Name'].unique())
otherApps

exclude = ['Liquity']
for item in exclude : 
    otherApps.remove(item)
    
compDf = pd.DataFrame({'Name' : otherApps})
redeemers = list(redemptions['account'].unique())   
```

```python
def compTVL(row):
    return userDf[userDf.Name == row['Name']]['Value'].sum()

def compUsers(row):
    return len(userDf[userDf.Name == row['Name']])

def AmountRedeemed(row): 
    userList = list(redemptions['account'].unique()) 
    return redemptions[(redemptions.account.isin(userList))]['lusdAmount'].sum()
    
def lostUsers(row):
    return len(userDf[userDf.Name == row['Name']])
            
    
compDf['TVL'] =  compDf.apply(compTVL, axis=1)      
compDf['nUsers'] =  compDf.apply(compUsers, axis=1)      
#compDf['AmountRedeemed'] =  compDf.apply(AmountRedeemed, axis=1) 
compDf['lostUsers'] =  compDf.apply(lostUsers, axis=1) 
```
